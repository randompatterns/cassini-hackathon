```python
from PIL import Image
import os
import rasterio
from matplotlib import pyplot as plt
```

```python
def make_gif(path):
    images = sorted(
        [os.path.join(path, x) for x in os.listdir(path) if x.endswith('.png')]
    )
    frames = [Image.open(x) for x in images]
    frame_one = frames[0]
    frame_one.save("addis_ababa.gif", format="GIF", append_images=frames,
               save_all=True, duration=750, loop=1)
```

```python
path = 'addis_ababa_gif/'
tif_images = [x for x in os.listdir(path) if x.endswith('.tif')]
```

```python
for ti in tif_images:
    with rasterio.open(os.path.join(path, ti)) as f:
        img = f.read()
        year = ti.split('_')[1].split('-')[0]
        plt.title(f'Addis Ababa: {year}')
        plt.imshow(img[0], cmap='gray')
        plt.axis('off')
        plt.savefig(os.path.join(path, ti.replace('.tif', '.png')),
                    bbox_inches='tight')
```
